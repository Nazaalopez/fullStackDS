```python
# Imports
import pandas as pd
import os
import numpy as np
```

```python
# Load CSV files in data/raw and read them as dataframes
data_dir = '../../data/raw'
csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

dataframes = {}
for file in csv_files:
    file_path = os.path.join(data_dir, file)
    df_name = os.path.splitext(file)[0]
    dataframes[df_name] = pd.read_csv(file_path)

# Display the loaded dataframes
for name, df in dataframes.items():
    print(f"DataFrame: {name}")
    display(df)
```

```python
# Clean data dictionary
dataframes['data_dic'] = dataframes.pop('Data Dictionary')
data_dic = dataframes['data_dic']
data_dic.rename(columns=data_dic.iloc[1], inplace = True)
data_dic = data_dic.iloc[2:, 1:]
data_dic
```

```python
# Review dataframes info
for name, df in dataframes.items():
    print(f"DataFrame: {name}")
    df.info()
    print("\n")
```

Notes:
- The `fact_txn` dataframe has 115,274 rows and 7 columns, containing transaction data.
- - physician_id and txn_location_type have null values, some datatypes shall be changed
- The `dim_patient` dataframe has 4,020 rows and 3 columns, containing patient demographic information.
- - non null values, requires datatypes changes 
- The `dim_physician` dataframe has 25,343 rows and 5 columns, containing physician demographic information.
- - gender and birth_year have some null values, datatypes changes
- The `model_table` dataframe has 2 columns, containing model-related data.
- The `data_dic` dataframe provides metadata for the other dataframes, including column names, data types, and descriptions.

```python
# Clean and transform fact_txn dataframe
fact_txn = dataframes['fact_txn']

# Handle missing values
#fact_txn['PHYSICIAN_ID'].fillna(-1, inplace=True)  # Fill missing physician_id with -1
#fact_txn.fillna({'TXN_LOCATION_TYPE': 'Unknown'}, inplace=True) # Fill missing txn_location_type with 'Unknown'

# Convert data types
# Datetime
fact_txn['TXN_DT'] = pd.to_datetime(fact_txn['TXN_DT'])  # Convert TXN_DT to datetime
# Numeric
#errors='coerce' invalid values will be set as NaN, Int64 will convert NaN to NA pandas
fact_txn['PATIENT_ID'] = pd.to_numeric(fact_txn['PATIENT_ID'], downcast='signed', errors='coerce').astype('Int64')
fact_txn['PHYSICIAN_ID'] = pd.to_numeric(fact_txn['PHYSICIAN_ID'], downcast='signed', errors='coerce').astype('Int64')
# Strings
fact_txn['PATIENT_ID'] = fact_txn['PATIENT_ID'].astype(str)  # Convert PATIENT_ID to string
fact_txn['PHYSICIAN_ID'] = fact_txn['PHYSICIAN_ID'].astype(str)  # Convert PHYSICIAN_ID to string
# Categories
fact_txn['TXN_LOCATION_TYPE'] = fact_txn['TXN_LOCATION_TYPE'].astype('category')
fact_txn['INSURANCE_TYPE'] = fact_txn['INSURANCE_TYPE'].astype('category')
fact_txn['TXN_TYPE'] = fact_txn['TXN_TYPE'].astype('category')
fact_txn['TXN_DESC'] = fact_txn['TXN_DESC'].astype('category')

# Rename columns
fact_txn.rename(columns={'PHYSICIAN_ID': 'HCP_ID'}, inplace=True)

# Display the cleaned and transformed dataframe
fact_txn.info()
fact_txn.head()
```

```python
# Apply upper case to TXN_DESC column to avoid missmatches 
fact_txn['TXN_DESC'] = fact_txn ['TXN_DESC'].str.upper()
fact_txn.head()
```

```python
# Clean and transform dim_patient dataframe
dim_patient = dataframes['dim_patient']

# Convert data types
# Numeric
dim_patient['PATIENT_ID'] = pd.to_numeric(dim_patient['PATIENT_ID'], downcast='signed', errors='coerce').astype('Int64')
dim_patient['BIRTH_YEAR'] = pd.to_numeric(dim_patient['BIRTH_YEAR'], downcast='signed', errors='coerce').astype('Int64')

# Strings
dim_patient['PATIENT_ID'] = dim_patient['PATIENT_ID'].astype(str)  # Convert PATIENT_ID to string

# Categories
dim_patient['GENDER'] = dim_patient['GENDER'].astype('category')

# Rename columns
dim_patient.rename(columns={'BIRTH_DATE': 'BIRTH_DT'}, inplace=True)


# Display the cleaned and transformed dataframe
dim_patient.info()
dim_patient.head()
```

```python
# Clean and transform dim_physician dataframe
dim_physician = dataframes['dim_physician']

# Handle missing values if necessary (e.g., fill with a placeholder or drop)
# dim_physician['GENDER'].fillna('Unknown', inplace=True)
# dim_physician['BIRTH_YEAR'].fillna(-1, inplace=True)

# Convert data types
# Numeric
dim_physician['PHYSICIAN_ID'] = pd.to_numeric(dim_physician['PHYSICIAN_ID'], downcast='signed', errors='coerce').astype('Int64')
dim_physician['BIRTH_YEAR'] = pd.to_numeric(dim_physician['BIRTH_YEAR'], downcast='signed', errors='coerce').astype('Int64')

# Strings
dim_physician['PHYSICIAN_ID'] = dim_physician['PHYSICIAN_ID'].astype(str)  # Convert PHYSICIAN_ID to string

# Categories
dim_physician['STATE'] = dim_physician['STATE'].astype('category')
dim_physician['PHYSICIAN_TYPE'] = dim_physician['PHYSICIAN_TYPE'].astype('category')
dim_physician['GENDER'] = dim_physician['GENDER'].astype('category')

# Rename columns
dim_physician.rename(columns={'PHYSICIAN_TYPE': 'HCP_SPECIALTY'}, inplace=True)
dim_physician.rename(columns={'PHYSICIAN_ID': 'HCP_ID'}, inplace=True)
dim_physician.rename(columns={'BIRTH_DATE': 'BIRTH_DT'}, inplace=True)

# Display the cleaned and transformed dataframe
dim_physician.info()
dim_physician.head()
```

```python
# Age calculator function 
from datetime import date
def calculate_age(birth):
    today = date.today()
    return today.year - birth.year - ((today.month, today.day) < (birth.month, birth.day))
```

```python
dim_physician['BIRTH_DT'] = pd.to_datetime(dim_physician['BIRTH_YEAR'], format = '%Y')
dim_patient['BIRTH_DT'] = pd.to_datetime(dim_patient['BIRTH_YEAR'], format = '%Y')
```

```python
# Calculate current age (aprox.) of the patient & hcp as a new column
dim_patient['PATIENT_AGE'] = dim_patient['BIRTH_DT'].apply(calculate_age)
dim_patient.head()
```

```python
# Calculate current age (aprox.) of the patient & hcp as a new column
dim_physician['HCP_AGE'] = dim_physician['BIRTH_DT'].apply(calculate_age)
dim_physician['HCP_AGE'] = pd.to_numeric(dim_physician['HCP_AGE'], downcast='signed', errors='coerce').astype('Int64')
dim_physician.head()
```

```python
# Format PATIENT_GENDER as expected
def gender_formater(value):
    if value == 'M':
        return 'M-Male'
    elif value == 'F':
        return 'F-Female'
    else:
        return 'U-Unknown'
```

```python
dim_patient['PATIENT_GENDER'] = dim_patient['GENDER'].map(gender_formater)
dim_patient.head(10)
```

```python
dim_physician['HCP_GENDER'] = dim_physician['GENDER'].map(gender_formater)
dim_physician.head(10)
```

```python
# Get state name from abbreviation from dim_physician
import us 
state_dict = us.states.mapping('abbr', 'name')
dim_physician['STATE_NAME'] = dim_physician['STATE'].map(state_dict)
dim_physician['STATE_NAME'] = dim_physician['STATE_NAME'].astype('category')
dim_physician.head()
```

```python
dim_physician.drop(columns=['BIRTH_DT'], inplace=True)
#dim_patient.drop(columns=['BIRTH_DT'], inplace=True)
dim_physician.drop(columns=['BIRTH_YEAR'], inplace=True)
dim_patient.drop(columns=['BIRTH_YEAR'], inplace=True)
dim_physician.drop(columns=['GENDER'], inplace=True)
dim_patient.drop(columns=['GENDER'], inplace=True)
```

```python
# Define the directory to save the intermediate CSV files
intermediate_dir = '../../data/intermediate'

# Ensure the directory exists
os.makedirs(intermediate_dir, exist_ok=True)

# Save each dataframe to a CSV file in the intermediate directory
for name, df in dataframes.items():
    file_path = os.path.join(intermediate_dir, f"{name}.csv")
    df.to_csv(file_path, index=False)
    print(f"Saved {name} to {file_path}")
```
